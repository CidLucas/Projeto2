```python
import pandas as pd
import numpy as np
```

```python
lista_data = []
for moeda in lista_moedas:
    operacoes = cg.get_coin_ticker_by_id(id = moeda)['tickers']
    moeda = []
    preco = []
    mercado = []
    confiavel = []
    hora = []
    
    for item in operacoes:
        if item['target'] == 'USDT' or item['target'] =='USD':
            moeda.append(item['base'])
            mercado.append(item['market']['name'])
            preco.append(item['converted_last']['usd'])
            confiavel.append(item['trust_score'])
            hora.append(item['timestamp'])
        
    tabela_raw = pd.DataFrame(list(zip(moeda, mercado, preco, confiavel, hora)), 
                              columns = ['ativo', 'exchange', 'preco', 'trust_score', 'timestamp'])
    lista_data.append(tabela_raw)

```

```python
print(type(preco[0]))
```

```python
df = pd.concat(lista_data, axis = 0)
```

```python
df
```

```python

preco_medio = df.groupby('ativo', as_index=False).agg({'preco':'mean'})
preco_medio.columns = ['ativo', 'preco_medio']
 
df = pd.merge(df, preco_medio, on='ativo')
df['Desvio'] = ((df['preco'] - df['preco_medio'])*100/df['preco_medio'])


# por que aparecem estas colunas? Por que a coluna desvio não aparece na base de dados? 
```

```python
df = df.drop_duplicates(subset = ['ativo', 'exchange'])
```
